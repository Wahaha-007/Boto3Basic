### 1. Get overall 100 repos

```python
import json
repos = json.loads(res.content.decode('utf-8'))
```

```python
type(repos)
```

```python
for repo in repos:
    print({'id': repo['id'], 'owner.login': repo['owner']['login'], 'name': repo['name']})
```
